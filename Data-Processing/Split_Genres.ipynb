In questo notebook vengono calcolate la deviazione standard, la media, e il numero di voti totali per i vari generi.

```python
import pandas as pd
import numpy as np
```

```python
dfr = pd.read_csv(r"C:\Users\giorgio1\Dropbox\Data Management\dataframes_definitivi/rotten_count.csv")
dfm = pd.read_csv(r"C:\Users\giorgio1\Dropbox\Data Management\dataframes_definitivi\metac_sent.csv")
```

```python
dfr = dfr.sort_values("count_rotten", ascending=False)
dfr.drop_duplicates(subset="Movie", inplace=True)
```

```python
dfr["genre"] = dfr["genre"].str.replace(' ', '')
dfm["genre"] = dfm["genre"].str.replace(' ', '')
```

```python
dfr.dropna(subset = ["genre"], inplace = True, axis = 0)
dfm.dropna(subset = ["genre"], inplace = True, axis = 0)
```

```python
x = dfr.apply(lambda x: x["genre"].split(","), axis=1)
y = dfm.apply(lambda x: x["genre"].split(","), axis=1)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
mlb = MultiLabelBinarizer()
array_out = mlb.fit_transform(x)
array_out2 = mlb.fit_transform(y)
```

```python
df_out = pd.DataFrame(data=array_out, columns=mlb.classes_)
df_out2 = pd.DataFrame(data=array_out2, columns=mlb.classes_)
```

```python
dfr = dfr.merge(df_out, right_on = dfr.index, left_on = df_out.index)
dfm = dfm.merge(df_out2, right_on = dfm.index, left_on = df_out2.index)
```

```python
dfr.drop(["key_0"], axis=1, inplace=True)
dfm.drop(["key_0"], axis=1, inplace=True)
```

```python
dfr.columns
```

```python
dfm = dfm.dropna(subset=['std_met'], axis = 0)
```

```python
dfr=dfr[dfr['count_rotten']!=0]
```

```python
dfm.columns
```

```python
conta = 0
diz = {}
for i in dfr.columns:
    conta = conta +1
    if (conta >16):
        print(dfr[dfr[i]==1]['diff'].std())
```

```python
conta = 0
diz = {}
for i in dfr.columns:
    conta = conta +1
    if (conta >16):
        dz = {}
        dfappoggio = dfr[dfr[i]==1][['user_avg_rate', 'count', 'StdDev', 'rating', 'count_rotten', 'std_rott', 'diff']]
        dfappoggio2 = dfm[dfm[i]==1]
        media = dfappoggio['user_avg_rate'] * dfappoggio['count']
        stdv = dfappoggio['StdDev'] * dfappoggio['count']
        media_rot = dfappoggio['rating'] * dfappoggio['count_rotten']
        stdv_rot = dfappoggio['std_rott']*dfappoggio['count_rotten']
        media_met = dfappoggio2['metascore'] * dfappoggio2['tot_voti_met']
        stdv_met = dfappoggio2['std_met'] * dfappoggio2['tot_voti_met']
        dz['user_avg'] = sum(media)/sum (dfappoggio['count'])
        dz['user_stv'] = sum(stdv)/sum (dfappoggio['count'])
        dz['user_count'] = sum(dfappoggio['count'])
        dz['rotten_avg'] = sum(media_rot)/sum(dfappoggio['count_rotten'])
        dz['rotten_stv'] = sum(stdv_rot)/sum(dfappoggio['count_rotten'])
        dz['rotten_count'] = sum(dfappoggio['count_rotten'])
        dz['meta_avg'] = sum(media_met)/sum(dfappoggio2['tot_voti_met'])
        dz['meta_stv'] = sum(stdv_met) / sum(dfappoggio2['tot_voti_met'])
        dz['meta_count'] = sum(dfappoggio2['tot_voti_met'])
        dz['stdv_diff_user_rot'] = dfappoggio['diff'].std()
        dz['stdv_diff_user_met'] = dfappoggio2['diff'].std()
        diz[i] = dz
```

```python
diz
```

```python
df_output = pd.DataFrame.from_dict(diz, orient='index').round(2)
```

```python
df_output = df_output.reset_index()
```

```python
df_output = df_output.rename(columns={'index': 'genre'})
```
